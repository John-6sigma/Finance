```python
'''
John An
6-16-2021
Amortization Table
'''

import numpy_financial as npf

# Inputs:
startBalance = float(input("Enter the Investment amount:"))
period = int(input("Enter the number of years:"))
rate = int(input("Enter the APR as a %:"))

# Convert rate to a Percent
rate = ((rate / 100) / 12)

# Convert years into months
period = period * 12

#Initialize the accumulator for the intrest
totalInterest = 0.0

#Payment
payment = (npf.pmt(rate, period, startBalance) * -1)

#Ending Balance
endBalance = startBalance

# Display the header for the Table
print("%6s%18s%20s%22s%24s%25s" % ("Period", "Starting balance", "Principal", "Interest", "Total Payment", "Ending Balance"))

#Compute and display the results for each year
for year in range(1, period + 1):
      #payment = (startBalance * (rate)) / 1 - (1 + rate) ** period
      interest = startBalance * rate
      principle = payment - interest
      endBalance = startBalance - principle
      print("%4d%18.2f%20.2f%22.2f%24.2f%26.2f" % (year, startBalance, principle, interest, payment, endBalance))
      startBalance = endBalance
      totalInterest += interest

# Display the totals for the Period
print("ending Balance: $%0.2f" % endBalance)
print("Total interest paid: $%0.2f" % totalInterest)
```
